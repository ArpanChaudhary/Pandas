```python
## Import pandas and NumPy libraries
```

```python
import numpy as np
import pandas as pd
```

```python
## defin a function for the salary hike
```

```python
def salary_hike(x):
  if x > 15000:
    return np.ceil(x + (x * 0.15))
  else:
    return np.ceil(x + (x * 0.35))
```

```python
## create a dataset for applying the predefined function (salary_hike)
```

```python
df = pd.DataFrame({'Department':['Testing','Developing','FrontEnd'],'Employee':['Annie','John','Boyka'],
                   'Salary':[11500,15500,13500]})

df
```

```python
df['New_Salary'] = df['Salary'].apply(salary_hike)

df
```

```python
## apply len finction to the employee column
```

```python
df['Name_Length'] = df['Employee'].apply(len)

df
```

```python
## apply lambda function (sqrt to the salary column)
```

```python
df['Salary_sqrt'] = df['Salary'].apply(lambda x: np.sqrt(x))

df
```
