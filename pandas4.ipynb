```python
## Import pandas and NumPy libraries
```

```python
import numpy as np
import pandas as pd
```

```python
# create a dataframe with the missing values
```

```python
raw_data = {'First_name':['Jason',np.nan,'Tina','Jake','Amy','Madhu'],
            'Last_name':['Miller',np.nan,'Ali','Milner','Cooze','Patil'],
            'Age':[42,np.nan,36,24,64,25],
            'Sex':['M',np.nan,'F','M','F','F'],
            'Unit_1':[4,np.nan,np.nan,2,3,4],
            'Unit_2':[3,np.nan,np.nan,4,4,4]}
df = pd.DataFrame(raw_data,columns=['First_name','Last_name','Age','Sex','Unit_1','Unit_2'])
df
```

```python
# drop missing values
```

```python
df_no_missing = df.dropna()

df_no_missing
```

```python
# if any row having all null entries, it will get deleted as using 'how' method
```

```python
df_cleaned = df.dropna(how='all')

df_cleaned
```

```python
df
```

```python
# if any column having all null entries
```

```python
df['Final_Score'] = np.nan
```

```python
df
```

```python
df_cleaned_1 = df.dropna(axis=1,how='all')

df_cleaned_1
```

```python
# fill null values
```

```python
df_fillna = df.fillna(0)

df_fillna
```

```python
df
```

```python
# filling null values in Unit_1
```

```python
df['Unit_1'] = df['Unit_1'].fillna(df['Unit_1'].mean())

df
```

```python
# filling null values in Unit_2 based on the sex
```

```python
df['Unit_2'] = df['Unit_2'].fillna(df.groupby('Sex')['Unit_2'].transform('mean'))

df
```

```python
# in above execution, to fill null values is required according to the sex as grouped according to sex,
# so it took the female values then take mean value and fill in the null values.
```

```python
# bfill  or back-fill
# same data is filled in null values to the previous cell

df_bfill = df.fillna(method='bfill')

df_bfill
```

```python
# ffill  or forward-fill
# same data is filled in null values to the next cell

df_ffill = df.fillna(method='ffill')

df_ffill
```

```python
df['Final_Score'] = df['Final_Score'].fillna(df['Unit_1'] + df['Unit_2'])

df
```

```python
raw_data = {'Company':['Google','Google','Google','Google','Facebook','Facebook','Facebook','Facebook','Amazon','Amazon','Amazon','Amazon'],
            'Project':['1st','1st','2nd','2nd','1st','1st','2nd','2nd','1st','1st','2nd','2nd'],
            'Incharge':['Sunder','Ruth','Benjamin','Sergey','Mark','Shrey','Eduardo','David','Jeffray','Andy','Keith','Edith'],
            'Sales(in millions)':[54,24,31,12,23,24,34,31,32,23,52,13],
            'Profit(in millions)':[14,8,9,4,10,11,14,13,12,7,22,4]}

df = pd.DataFrame(raw_data,columns=['Company','Project','Incharge','Sales(in millions)','Profit(in millions)'])

df
```
