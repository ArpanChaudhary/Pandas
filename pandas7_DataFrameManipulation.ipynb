## DataFrame Manipulation

```python
matrix_data = np.matrix('22,66,140;42,70,148;30,62,125;35,68,160;25,62,152')

row_labels = ['Raju','Pandey','Srini','Sunil','Ashok']

column_headings = ['Age','Weight','Height']
```

```python
df = pd.DataFrame(data=matrix_data,index=row_labels,columns=column_headings)

print('\nA new DataFrame\n','-'*25, sep='')

print(df)
```

```python
print("\nThe 'Height' column\n", '-'*25, sep='')
print(df['Height'])
```

```python
print("\nThe 'Height' and 'Weight' columns indexed by passing a list\n", '-'*40, sep='')

print(df[['Height','Weight']])
```

```python
only_ht_wt = df[['Height','Weight']]

only_ht_wt
```

```python
print("\nThe 'Age' column accessed by DOT method(not recommended)\n",'-'*25, sep='')
print(df.Age)
```

```python
df.loc['Raju']   #fetching element using Index name
```

```python
df.iloc[0]      #fetching element using Index number
```

```python
print(df)
```

```python
print(df.loc['Pandey','Height'])
```

```python
print(df.loc['Ashok',['Height','Weight']])
```

```python
print(df.loc[['Ashok','Pandey'],['Height','Weight']])
```

```python
## Fetching Data based on some condition
```

```python
print("\nHeight > 150 inches\n",'-'*25, sep='')

print(df[df['Height']> 150])
```

```python
print("\nWeight > 60 lbs\n",'-'*25, sep='')

print(df[df['Weight']> 60])
```

```python
bool1 = df['Height']>140

bool2 = df['Weight']>60


print(df[(bool1) & (bool2)])
```

```python
print(df[bool1][['Weight','Height']])
```

```python
selected = df[bool1][['Weight','Height']]
```
