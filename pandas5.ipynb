```python
import numpy as np
import pandas as pd
```

## Data Manipulation

```python
df = pd.read_csv('/content/titanic_train.csv')
```

```python
df.head()
```

## How to check the unique values of any column

```python
df['Survived'].unique()
```

## How to rename categories using pandas

```python
new_survived = pd.Categorical(df['Survived'])

new_survived = new_survived.rename_categories(['Died','Survived'])

new_survived.describe()
```

```python
df['Pclass'].unique()
```

```python
new_Pclass = pd.Categorical(df['Pclass'])

new_Pclass = new_Pclass.rename_categories(['Class1','Class2','Class3'])

new_Pclass.describe()
```

```python
## re-assigning the column 'Pclass' with new values
```

```python
df['Pclass'] = new_Pclass
```

```python
df
```

```python
# Let's check for the age missing
```

```python
df['Age'].isna().sum()
```

```python
age_missing = df[~df.Age.isin(['NaN'])]

age_missing
```

## Find the maximum fare

```python
maximum_fare = df[df.Fare.isin([max(df['Fare'])])]

maximum_fare
```

## Find the minimum fare

```python
minimum_fare = df[df.Fare.isin([min(df['Fare'])])]

minimum_fare
```
